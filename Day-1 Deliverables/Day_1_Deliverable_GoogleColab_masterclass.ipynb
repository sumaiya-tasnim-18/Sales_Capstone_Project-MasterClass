Exercise 2: Mathematical Operations

```python
# Creating variables
product_price = 120.50
product_quantity = 3
tax_rate = 0.08  # 8%

# Calculations
subtotal = product_price * product_quantity
tax_amount = subtotal * tax_rate
total_cost = subtotal + tax_amount

# Displaying results
print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax Amount: ${tax_amount:.2f}")
print(f"Total Cost: ${total_cost:.2f}")

# Sales comparison
sales_period1 = 1500
sales_period2 = 1800

# Calculating percentage increase
percentage_increase = ((sales_period2 - sales_period1) / sales_period1) * 100

# Displaying percentage increase formatted to two decimal places
print(f"Percentage Increase in Sales: {percentage_increase:.2f}%")
```

Exercise 3: Lists

```python
# Creating the list of monthly expenses
monthly_expenses = [2500, 3000, 2800, 3200, 2900]

# Calculating total, average, minimum, and maximum expenses
total_expenses = sum(monthly_expenses)
average_expense = total_expenses / len(monthly_expenses)
minimum_expense = min(monthly_expenses)
maximum_expense = max(monthly_expenses)

# Displaying the calculated results
print(f"Total Expenses: ${total_expenses}")
print(f"Average Expense: ${average_expense:.2f}")
print(f"Minimum Expense: ${minimum_expense}")
print(f"Maximum Expense: ${maximum_expense}")

# Adding a new expense to the list
monthly_expenses.append(3100)

# Recalculating total and average after adding the new expense
new_total_expenses = sum(monthly_expenses)
new_average_expense = new_total_expenses / len(monthly_expenses)

# Displaying the updated results
print("\nAfter Adding a New Expense:")
print(f"New Total Expenses: ${new_total_expenses}")
print(f"New Average Expense: ${new_average_expense:.2f}")
```

Exercise 5: Conditionals

```python
# Defining a function to check discount eligibility
def check_discount(purchase_amount):
    # Checking purchase amount and applying conditions
    if purchase_amount > 500:
        print("Eligible for 20% discount")
    elif purchase_amount > 100:
        print("Eligible for 10% discount")
    else:
        print("No discount")

# Calling the function with different purchase amounts to test all conditions
print("Purchase Amount: $50 →", end=" ")
check_discount(50)

print("Purchase Amount: $200 →", end=" ")
check_discount(200)

print("Purchase Amount: $800 →", end=" ")
check_discount(800)
```

Exercise 7: Pandas DataFrame Creation and Manipulation

```python
# Importing the pandas library
import pandas as pd

# Creating the student data dictionary
student_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math_Score': [85, 92, 78, 88],
    'Science_Score': [90, 88, 85, 95],
    'English_Score': [75, 80, 82, 79]
}

# Creating a DataFrame from the dictionary
students_df = pd.DataFrame(student_data)

# Displaying the entire DataFrame
print("Original Student DataFrame:")
print(students_df)

# Adding a new column by calculating the average score for each student
students_df['Average_Score'] = (
    students_df['Math_Score'] +
    students_df['Science_Score'] +
    students_df['English_Score']
) / 3

# Displaying the DataFrame after adding the new column
print("\nDataFrame with Average_Score column:")
print(students_df)

# Filtering the DataFrame to show students with Average_Score > 85
high_achievers = students_df[students_df['Average_Score'] > 85]

# Displaying the filtered DataFrame
print("\nStudents with Average_Score greater than 85:")
print(high_achievers)
```

Exercise 8: NumPy Array Operations

```python
# Importing the numpy library
import numpy as np

# Creating NumPy arrays for product sales
product_a_sales = np.array([120, 150, 130, 180, 160])
product_b_sales = np.array([100, 130, 140, 170, 150])

# Calculating total sales per month by adding arrays element-wise
total_sales_per_month = product_a_sales + product_b_sales
print("Total Sales per Month:", total_sales_per_month)

# Calculating the difference in sales (Product A - Product B)
difference_in_sales = product_a_sales - product_b_sales
print("Difference in Sales (A - B):", difference_in_sales)

# Calculating average sales for both products
average_sales_a = np.mean(product_a_sales)
average_sales_b = np.mean(product_b_sales)
print(f"Average Sales of Product A: {average_sales_a}")
print(f"Average Sales of Product B: {average_sales_b}")

# Creating a boolean array to check where Product A sales > Product B sales
is_a_greater = product_a_sales > product_b_sales
print("Is Product A Greater Than Product B Each Month?:", is_a_greater)

# Counting how many months Product A had greater sales
count_a_greater = np.sum(is_a_greater)
print(f"Number of Months Product A Had Higher Sales: {count_a_greater}")
```

Exercise 9: Matplotlib Basic Plotting

```python
# Importing the matplotlib.pyplot module
import matplotlib.pyplot as plt

# Creating data for the line plot
x_values = [1, 2, 3, 4, 5]
y_values = [2, 4, 5, 4, 5]

# Creating a simple line plot
plt.plot(x_values, y_values, label="Data Series", marker='o', color='b')

# Adding title and axis labels
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Adding legend
plt.legend()

# Displaying the line plot
plt.show()

# Creating data for the bar chart
categories = ['A', 'B', 'C']
values = [10, 15, 7]

# Creating a bar chart
plt.bar(categories, values, color='pink')

# Adding title and axis labels
plt.title("Category-wise Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")

# Displaying the bar chart
plt.show()
```

Exercise 10: Seaborn Basic Plotting

```python
# Importing required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Creating a DataFrame for category data
category_data = pd.DataFrame({
    'Category': ['Type A', 'Type B', 'Type A', 'Type C', 'Type B', 'Type C'],
    'Value': [10, 15, 12, 8, 18, 10]
})

# Creating a count plot for the 'Category' column
sns.countplot(data=category_data, x='Category', palette='pastel')
plt.title("Count of Each Category")

# Displaying the count plot
plt.show()

# Creating a bar plot showing average 'Value' per category
sns.barplot(data=category_data, x='Category', y='Value', estimator=np.mean, palette='Set2')
plt.title("Average Value per Category")

# Displaying the bar plot
plt.show()

# Creating a small correlation matrix using NumPy
correlation_matrix = np.array([
    [1.0, 0.8, 0.5],
    [0.8, 1.0, 0.6],
    [0.5, 0.6, 1.0]
])

# Displaying the correlation matrix as a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")

# Displaying the heatmap
plt.show()
```
